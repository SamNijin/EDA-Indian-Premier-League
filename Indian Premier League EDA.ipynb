```python
def displayColumns(dataset):
    columns=list(dataset.columns)
    return display(pd.DataFrame(columns,columns=["Columns"]))
```

```python
def count_plot_(data, y, limit, text, op_val=True):
    fig, (ax1, ax2) = plt.subplots(2,)

    fig.tight_layout(pad=2)

    fig.add_subplot(sns.countplot(data=data, y=y, ax=ax1))

    fig.add_subplot(sns.countplot(data=data, y=y, ax=ax2))
    plt.xlim(limit)

    plt.show()
    
    
    text = text.split('-')

    x = [i for i in deliveries_df[y].value_counts()]
    
    
    for i in range(len(x)):
        data = x[i]
        text.insert(1, data)
        if op_val:
            text.append(' '+str(i+1))
        
        
        for i in text:
            text_ = ''.join(str(i))
            print(text_,end='')
              
        print()
        
        del text[1]
        if op_val:
            del text[-1]
```
